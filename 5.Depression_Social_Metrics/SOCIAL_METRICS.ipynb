```python
import pandas as pd
df = pd.read_csv('intent_mod.csv')
df.head(10)
```

```python

df['Depressed'] = df['Depressed'].replace({'NO': 0 ,'YES' : 1})
```

```python
y_true = df['Depressed'].tolist()
y_pred = df['Sentiment_score'].abs().tolist()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import roc_curve, auc, cohen_kappa_score

# Generate some example data

# Compute ROC curve and AUC score
fpr, tpr, thresholds = roc_curve(y_true, y_pred)
roc_auc = auc(fpr, tpr)

# Compute Kappa score
kappa = cohen_kappa_score(y_true, np.round(y_pred))

# Create figure with dark theme
fig, ax = plt.subplots(facecolor='#2b2b2b')

# Add ROC curve to plot
ax.plot(fpr, tpr, color='r', lw=2, label='ROC curve (AUC = %0.2f)' % roc_auc)

# Add random classifier line to plot
ax.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')

# Set axis labels and title
ax.set_xlabel('False Positive Rate', color='w')
ax.set_ylabel('True Positive Rate', color='w')
ax.set_title('Receiver operating characteristic (ROC) curve', color='w')

# Add legend
ax.legend(loc="lower right")

# Set background color and grid color
ax.set_facecolor('#ffffff')
ax.grid(color='grey', linestyle='--', alpha=0.5)

# Add kappa score text to plot
plt.text(0.75, 0.25, 'Kappa score = %0.2f' % kappa, color='w', fontsize=12,
         bbox=dict(facecolor='grey', alpha=0.5))

# Show plot
plt.show()
```
